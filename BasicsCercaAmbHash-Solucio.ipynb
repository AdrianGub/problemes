Aquest fitxer mostra les solucions als exercicis bàsics de cerca amb Hash [Basics Hash](BasicsCercaAmbHash.ipynb)

### Volem guardar emmagatzemar informació dels estudiants. Partim d'una classe amb 99 alumnes, i volem crear una estructura que ens permeti cercar un alumne ràpidament.

### Conceptualització del problema

- Quina estructura podem crear per tal que cada cerca tingui cost O(1)? (el problema no es pot resoldre amb una estructura pròpia de Python)

L'estructura triada serà una taula Hash. Caldrà prendre algunes decisions:

#### Decisió 1: mida de la taula

Com que tenim 98 alumnes farem una taula de 101 posicions(101 és un nombre primer).


#### Decisió 2: funció Hash

Usarem una funció hash universal. Així, representarem cada alumne com una tupla de tres enters mòdul 101 (els primers dos dígits quasi sempre són '18' i els descartem)

Per ex:
- niub18518012 => (51,80,12)

I definirem 3 coeficients mòdul 101

a => (13,47, 77)

```python
# la llista de NIUBs és la segÜent

llistaniubs = ['niub18518012', 'niub18589763', 'niub18589776', 'niub16973034',
               'niub19125491', 'niub18518486', 'niub18518330', 'niub18518500',
               'niub16347854', 'niub18588533', 'niub18566891', 'niub16888180',
               'niub18519741', 'niub18596201', 'niub18518114', 'niub18588572',
               'niub16625902', 'niub14730841', 'niub16584633', 'niub18519702',
               'niub18518511', 'niub18518450', 'niub18589754', 'niub18588522',
               'niub18518425', 'niub16440881', 'niub18518051', 'niub16043701',
               'niub18588114', 'niub16805093', 'niub18519763', 'niub16687712',
               'niub18589710', 'niub14763863', 'niub18596304', 'niub18598894',
               'niub18518006', 'niub18518442', 'niub18518091', 'niub16828450',
               'niub18588511', 'niub18588232', 'niub16831426', 'niub18518346',
               'niub11236310', 'niub18588120', 'niub18567770', 'niub10196663',
               'niub19121165', 'niub16815441', 'niub18589726', 'niub18589794',
               'niub18569595', 'niub16609841', 'niub18599174', 'niub18519794',
               'niub18518136', 'niub16803506', 'niub18569091', 'niub10308215',
               'niub16120940', 'niub16313033', 'niub16843842', 'niub18581933',
               'niub18588595', 'niub18597724', 'niub18595521', 'niub16759784',
               'niub18518383', 'niub18518464', 'niub16421613', 'niub16839025',
               'niub18518324', 'niub16766465', 'niub16271220', 'niub16699055',
               'niub18588184', 'niub16619210', 'niub18588544', 'niub18519780',
               'niub18588170', 'niub18518065', 'niub18588162', 'niub18519735',
               'niub18518040', 'niub18593112', 'niub18518305', 'niub18593525',
               'niub18588581', 'niub18567304', 'niub18588136', 'niub18518490',
               'niub19125804', 'niub18589735', 'niub18588566', 'niub16806162',
               'niub18518475', 'niub16638516', 'niub16633574']


#  els alumnes a buscar seran els següents:


import random


def llistaABuscar():
    abuscar = []
    for i in range(0, 1000):
        x = random.randint(0, 98)
        abuscar.append(llistaniubs[x])
    return abuscar


abuscar = llistaABuscar()

print(abuscar)
```

```python
# definim la funció hash:


def hash(niub):
    niub = int(niub[6:12])
    return (niub // 10000 * 13 + niub % 10000 // 100 * 47
            + niub % 100 * 77) % 101
```

#### Implementació de la taula Hash

```python
# Creem una taula amb 101 posicions
# i guardem els niubs a la posició que els toca a la taula
# com a valor, guardarem l'index de la llistaniubs
taulahash = [-1]*101  # inicialitzem la taula amb -1

colisions = 0
maxim = 0
posLlistaNIUB = 0  # posició inicial a llistaniubs
for niub in llistaniubs:
    index = hash(niub)
    if taulahash[index] == -1:         # encara no hi ha cap element
        taulahash[index] = [(niub, posLlistaNIUB)]
    else:
        taulahash[index].append((niub, posLlistaNIUB))
        # hi ha un o més elements
        colisions += 1
        maxim = max(maxim, len(taulahash[index]))
    posLlistaNIUB += 1
print(maxim, colisions)
```

```python
# Finalment agafarem 1000 niubs a l'atzar comptarem el temps de cerca
# amb hash i amb cerca binària.
# Usarem alguns dels algorismes que coneixem per cercar o per ordenar

import timeit


def recBinSearch(x, nums, low, high):
    """
    Algorisme bàsic de cerca binària recursiva
    :param x el nombre a trobar
    :param nums la llista de nums a trobar
    :param low l'índex més baix de la subllista on estem cercant ara
    :param high l'índex més alt de la subllista on estem cercant ara
    :return posició de x a nums o -1 si no s'ha trobar
    """
    if low > high:
        return -1
    mid = (low + high) // 2
    items = nums[mid]
    if items == x:
        return mid
    elif x < items:
        return recBinSearch(x, nums, low, mid-1)
    else:
        return recBinSearch(x, nums, mid+1, high)


def partition(A, first, last):
    """
    Aquesta funció tria el valor mig entre l'element inicial, final i mig
    d'una llista i l'ubica al lloc que li correspondrà quan la llista està
    ordenada, alhora deixa a la seva esquerra valors menors i a la dreta
    valors majors. Per tant altera A
    A més retorna la posició de la partició
    """
    mid = (first + last) // 2   # ordenem A[first],A[mid],A[last]
    if A[first] > A[mid]:
        A[first], A[mid] = A[mid], A[first]
    if A[first] > A[last]:
        A[first], A[last] = A[last], A[first]
    if A[mid] > A[last]:
        A[mid], A[last] = A[last], A[mid]
    A[mid], A[first] = A[first], A[mid]  # valor mig a l’inici
    pivot = first
    i = first + 1
    j = last
    while True:
        while i <= last and A[i] <= A[pivot]:
            i += 1
        while j >= first and A[j] > A[pivot]:
            j -= 1
        if i >= j:
            break
        else:
            A[i], A[j] = A[j], A[i]  # intercanviem, fem avançar i j
    A[j], A[pivot] = A[pivot], A[j]  # vector partit, pivot=j
    return j


def quick_sort(A):
    quick_sort_r(A, 0, len(A) - 1)


def quick_sort_r(A, first, last):
    if last > first:
        pivot = partition(A, first, last)
        quick_sort_r(A, first, pivot - 1)
        quick_sort_r(A, pivot + 1, last)


def cercaNostra(abuscar):
    posicions = []
    for n in abuscar:
        h = hash(n)                    # aquest és l'índex de la taula
        index = 0                      # posicio dins la llista hash
        while n != taulahash[h][index][0]:
            index += 1
        posicions.append(h)
        return posicions


def cercaBinaria(abuscar):
    posicions = []
    quick_sort(llistaniubs)
    for n in abuscar:
        index = recBinSearch(n, llistaniubs, 0, 98)
        posicions.append(index)
        return posicions
```
